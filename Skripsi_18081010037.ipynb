```python
tokennize=[]
for a in range(len(text)) :
    clean_angka=re.sub(r"\d+", "", text[a].lower())
    clean_tanda= clean_angka.translate(str.maketrans("","",string.punctuation))
    clean_whitespace= clean_tanda.strip()
    pisah = clean_whitespace.split()
    tokennize.append(pisah)
```

```python
def cal_entro_total():
    global sentimen_latih
    global kategori_0
    global kategori_1
    global kategori_2
    global kategori_3
    global kategori_4
    global kategori_5
    global kategori_6
    global entropy_total
    
    kategori_0=0
    kategori_1=0
    kategori_2=0
    kategori_3=0
    kategori_4=0
    kategori_5=0
    kategori_6=0
    hitung_0=0
    hitung_1=0
    hitung_2=0
    hitung_3=0
    hitung_4=0
    hitung_5=0
    hitung_6=0
    for a in range(len(sentimen_latih)):

        if sentimen_latih[a]==0:
            kategori_0+=1
        elif sentimen_latih[a]==1:
            kategori_1+=1
        elif sentimen_latih[a]==2:
            kategori_2+=1
        elif sentimen_latih[a]==3:
            kategori_3+=1
        elif sentimen_latih[a]==4:
            kategori_4+=1
        elif sentimen_latih[a]==5:
            kategori_5+=1
        elif sentimen_latih[a]==6:
            kategori_6+=1
            
    if kategori_0==0:
        hitung_0==0
    else:
        hitung_0=-kategori_0/len(sentimen_latih)*np.log2(kategori_0/len(sentimen_latih))
    
    if kategori_1==0: 
        hitung_1==0
    else:
        hitung_1=-kategori_1/len(sentimen_latih)*np.log2(kategori_1/len(sentimen_latih))
    
    if kategori_2==0: 
        hitung_2==0
    else:
        hitung_2=-kategori_2/len(sentimen_latih)*np.log2(kategori_2/len(sentimen_latih))
    
    if kategori_3==0:
        hitung_3==0
    else:
        hitung_3=-kategori_3/len(sentimen_latih)*np.log2(kategori_3/len(sentimen_latih))
    
    if kategori_4==0 :
        hitung_4==0
    else:
        hitung_4=-kategori_4/len(sentimen_latih)*np.log2(kategori_4/len(sentimen_latih))
    
    if kategori_5==0 :
        hitung_5==0
    else:
        hitung_5=-kategori_5/len(sentimen_latih)*np.log2(kategori_5/len(sentimen_latih))
    
    if kategori_6==0 :
        hitung_6==0
    else:
        hitung_6=-kategori_6/len(sentimen_latih)*np.log2(kategori_6/len(sentimen_latih))
    entropy_total=hitung_0+hitung_1+hitung_2+hitung_3+hitung_4+hitung_5+hitung_6
    
```
